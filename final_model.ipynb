```python
#import libraries
import pandas as pd

#import dataset
df_1 = pd.read_csv('results/final_submmit_tree_decision.csv', sep = ';', index_col = 0)
df_2 = pd.read_csv('results/final_submmit_polynomial_regression.csv', sep = ';', index_col = 0)
```

```python
#Calculate dif
dif = []
for key,line in df_2.iterrows():
    city = line.city_code
    year = line.year
    product = line['product']
    product_type = line.product_type
    aux_df = df_1.loc[df_1['city_code'] == city].copy()
    aux_df = aux_df.loc[aux_df['product'] == product]
    aux_df = aux_df.loc[aux_df['product_type'] == product_type]
    aux_df = aux_df.loc[aux_df['year'] == year]
    dif.append(abs(line.destinated_area - aux_df.destinated_area.values[0]))
```

```python
dif
```

```python
mean = []
for key,line in df_2.iterrows():
    city = line.city_code
    year = line.year
    product = line['product']
    product_type = line.product_type
    aux_df = df_1.loc[df_1['city_code'] == city].copy()
    aux_df = aux_df.loc[aux_df['product'] == product]
    aux_df = aux_df.loc[aux_df['product_type'] == product_type]
    aux_df = aux_df.loc[aux_df['year'] == year]
    mean.append(abs(line.destinated_area + aux_df.destinated_area.values[0])/2)

df_2.loc[:,'destinated_area'] = mean
```

```python
df_2.to_csv('results/8GDLJG.csv',sep = ';')
```
